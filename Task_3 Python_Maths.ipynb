### Problem 1 :  Integers Come In All Size

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())

eqn = (a**b)+(c**d)
print(eqn)
```

### Problem 2 : Polar Coordinates

```python
import cmath
z= input()
s = complex(z)
print(abs(s))
print(cmath.phase(s))
```

### Problem 3 : Fing Angle MBC

```python
from math import atan, degrees
AB = int(input())
BC = int(input())

angleC = atan(AB/BC)
print((round(degrees(angleC))),chr(176),sep='')
```

### Problem 4 : Mod Divmod

```python
a = int(input())
b = int(input())
print(a//b)
print(a%b)
print(divmod(a,b))
```

### Problem 5 : Power - Mod Power

```python
a = int(input())
b = int(input())
m = int(input())
print(pow(a,b),pow(a,b,m),sep='\n')
```
