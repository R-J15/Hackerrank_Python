###  Problem 3 : Nested Lists

```python
result = []
scorelist = []

if __name__ == '__main__':
    for _ in range(int(input())):
        name = input()
        score = float(input())
        
        result += [[name,score]]
        scorelist += [score]
        
    a = sorted(list(set(scorelist)))[1]
        
    for b,c in sorted(result):
        if c == a:
            print(b)
     
```

### Problem 4 : Find the percentage

```python
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    
    score = student_marks[query_name]
    avg = sum(score)/len(score)
    print("%.2f"%avg)
```

### Problem 5 : Lists

```python
N = int(input())
l = []
for i in range(N):
    s = list(input().split())
    if s[0]=='insert':
        l.insert(int(s[1]),int(s[2]))
    if s[0]=='remove':
        l.remove(int(s[1]))
    if s[0]=='append':
        l.append(int(s[1]))
    if s[0]=='sort':
        l.sort()
    if s[0]=='pop':
        l.pop()
    if s[0]=='reverse':
        l.reverse()     
    if s[0]=='print':
        print(l)
     
```

### Problem 7 : Introduction to Sets

```python
def average(array):
    my_set = set(array)
    avg = sum(my_set)/len(my_set)
    return(avg)

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```

### Problem 11 : Set.discard().remove() & pop()

### Problem 16 : Set Mutation

```python
A = int(input())
s = set(map(int, input().split()))
N = int(input())

for i in range(N):
    cd = input().split()
    if cd[0]=="intersection_update":
        s2 = set(map(int, input().split()))
        s.intersection_update(s2)
    elif cd[0]=="update":
        s2 = set(map(int, input().split()))
        s.update(s2)
    elif cd[0]=="symmetric_difference_update":
        s2 = set(map(int, input().split()))
        s.symmetric_difference_update(s2)
    elif cd[0]=="difference_update":
        s2 = set(map(int, input().split()))
        s.difference_update(s2)
        
print(sum(s))
```

### Problem 19 : Check Strict Superset
