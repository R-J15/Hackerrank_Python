### Problem 6: Write a function

```python
year = int(input())
def is_leap(year):
    leap = False
    
    if year%400 ==0:
        leap = True
    elif year%4==0 and year%100 !=0:
        leap = True
        
    return leap

print(is_leap(year))

```
